```python
# What is an Outlier?
# Outlier is a value that is very different from other values in the dataset.


import pandas as pd

# Sample data
data = {
    "Marks": [50, 55, 60, 52, 58, 59, 100]  # 100 is outlier
}

df = pd.DataFrame(data)

# Calculate Q1 and Q3
Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df["Marks"] < lower_limit) | (df["Marks"] > upper_limit)]

print("Outliers:")
print(outliers)
```

```python
# Using Z-Score Method
import numpy as np
from scipy import stats

data = [50, 55, 60, 52, 58, 59, 100]

z_scores = np.abs(stats.zscore(data))

outliers = np.where(z_scores > 3)

print("Outlier index:", outliers)
print("Outlier value:", np.array(data)[outliers])
```
