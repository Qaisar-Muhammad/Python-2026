# operation

```python
# basic operations with different variable types in Python
a = 10
b= 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b) 
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

```python
# for bodmas rule
result = (a + b) * (a - b) / b
print("Result with BODMAS:", result)
```

```python
a= 20
b= 30   
c= 40
result = a + b * c
print("Result without BODMAS:", result)
```

```python
# difficult bodmas
a= 10
b= 5

result = a + b * c - a / b + a ** 2
print("Difficult BODMAS Result:", result)
```
