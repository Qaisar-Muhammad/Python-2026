```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 

# Example: Line Chart
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line 1')
plt.title('Example Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({'x': x, 'y': y})
print(df)
```

```python
type(df)
```

```python
df.plot(x='x', y='y', kind='bar')
plt.show
```

```python
df.plot(x='x', y='y', kind='barh')
plt.show
```

```python
df.plot(x='x', y='y', kind='scatter')
plt.show
```

```python
df.plot(x='x', y='y', kind='line')
plt.show
```
