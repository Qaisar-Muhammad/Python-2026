```python
#loop is used to execute a block of code repeatedly until a certain condition is met. There are two types of loops in Python: for loop and while loop.
#for loop is used to iterate over a sequence (such as a list, tuple, or string) or other iterable object. The syntax of a for loop is:
#for variable in sequence:
#    # code to be executed
#while loop is used to execute a block of code as long as a certain condition is true. The syntax of a while loop is:
#while condition:
#    # code to be executed
```

```python
#for odd or even numbers from 1 to 10
for i in range(1, 11):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
        
```

```python
a= 0
if a% 2 == 0:
    print(f"{a} is even")
else:
    print(f"{a} is odd")
    
```

```python
# from user input is it odd or even 
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
    
```

```python
# for disscount of 10% if price is greater than 1000
price = float(input("Enter the price: "))
if price > 1000:
    discount = price * 0.1
    final_price = price - discount
    print(f"Final price after discount: {final_price}")
    
```

```python
#iterate over a list of numbers and print the square of each number
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(f"The square of {num} is {num**2}")   
```

```python
# iterate string 
string = "Hello"
for char in string:
    print(char)
```

```python
# iterate over dictionary
person = {"name": "Alice", "age": 30, "city": "New York"}
for key, value in person.items():
    print(f"{key}: {value}")
    
```

```python
# student data with name marks and grade in list of of dictionary and iterate over it
students = [
    {"name": "Alice", "marks": 85, "grade": "A"},
    {"name": "Bob", "marks": 75, "grade": "B"},
    {"name": "Charlie", "marks": 65, "grade": "C"}
]
for student in students:
    print(f"Name: {student['name']}, Marks: {student['marks']}, Grade: {student['grade']}") 
    
```

```python
#iterate over a range of a  values
for i in range(5):
    print(i)
```

```python
# comprehension of list with for loop
squares = [x**2 for x in range(10)]
print(squares)
```

```python
# comprehension of dictionary with for loop
squares_dict = {x: x**2 for x in range(10)}
print(squares_dict)
```
