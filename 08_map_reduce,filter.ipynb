```python
# # map
# #area of a circle 
# import math
# def areaa(r):
#     return math.pi*(r**2)


# radii=[1,2,3,4,5]
# areas=[]
# for r in radii:
#     a=areas(r)
#     areas.append(a)

import math

def areaa(r):
    return math.pi * (r**2)

radii = [1,2,3,4,5]
areas = []

for r in radii:
    a = areaa(r)
    areas.append(a)

print(areas)

```

```python
# map function 
list(map(areaa,radii))
```

```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32
celsius_to_fahrenheit(54)
```

```python
# now farheneit to celsius
# now farheneit to celsius
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5 / 9
fahrenheit_to_celsius(87)
```

```python
import statistics
# now filter function 

data=[1,2,3,4,5,6,7,8,8,6,54,]
avg=statistics.mean(data)
print(avg)
```

```python
filter(lambda x: x>avg ,data )
```

```python
list(filter(lambda x: x>avg ,data ))
```

```python
# reduce function
from functools import reduce

# Example: sum all elements in 'data'
a = reduce(lambda x, y: x + y, data)
print(a)

```

```python
# other exampple 
# reduce function is used to apply a function cumulatively to the items of a sequence, from left to right, so as to reduce the sequence to a single value.
# For example, reduce(lambda x, y: x + y, data) will sum all elements in 'data'.

result = reduce(lambda x, y: x * y, data)
print(result)
```

```python
# Find elements in 'data' greater than 'avg' using a loop (not filter)
greater_than_avg = []
for x in data:
    if x > avg:
        greater_than_avg.append(x)
print(greater_than_avg)

```
