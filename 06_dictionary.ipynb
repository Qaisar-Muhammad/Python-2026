```python
# dictionary is a collection of key-value pairs. It is defined using curly braces {} and each key is separated from its value by a colon (:).
# Dictionaries are mutable, meaning you can change their content after they are created. They are also
# unordered, which means that the items do not have a defined order, and they cannot be accessed by index.
# Dictionaries are commonly used to store and manage data that is associated with unique keys, such as
# user information, configuration settings, or any other data that can be represented as key-value pairs.
```

```python
dic= {"name": "Alice", "age": 30, "city": "New York"}
print(dic) # {'name': 'Alice', 'age': 30, 'city':
```

```python
print(dic["name"]) # Alice
print(dic["age"]) # 30
print(dic["city"]) # New York
```

```python
# replace value for a key in a dictionary
dic["age"] = 31
print(dic) # {'name': 'Alice', 'age': 31, 'city':
```

```python
#insert a new key-value pair in a dictionary
dic["country"] = "USA"
print(dic) # {'name': 'Alice', 'age': 31, 'city': 'New York', 'country': 'USA'}
```

```python
print(len(dic)) # 4
```

```python
# delete a key-value pair from a dictionary
del dic["city"]
print(dic) # {'name': 'Alice', 'age': 31, 'country':
```

```python
# sort a dictionary by keys
sorted_dic = dict(sorted(dic.items()))
print(sorted_dic) # {'age': 31, 'country': 'USA', 'name': 'Alice'}
```

```python
#value () method returns a list of all the values in the dictionary.
values = dic.values()
print(values) # dict_values([31, 'USA', 'Alice'])
```

```python
#keys() method returns a list of all the keys in the dictionary.
keys = dic.keys()
print(keys) # dict_keys(['name', 'age', 'country'])
```

```python
#update() method is used to update the dictionary with the elements from another dictionary object or from an iterable of key-value pairs.
new_dic = {"name": "Bob", "age": 25}
dic.update(new_dic)
print(dic) # {'name': 'Bob', 'age': 25, 'country': 'USA'}
```
