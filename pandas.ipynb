```python
# Importing pandas library
import pandas as pd

# Creating two sample DataFrames for demonstration
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5],
    'Name': ['Charlie', 'David', 'Eve'],
    'Salary': [70000, 80000, 90000]
})

# Inner Join: Returns rows where there is a match in both DataFrames
inner_join = pd.merge(df1, df2, on='ID', how='inner')
# This will merge rows with matching 'ID' values
print("Inner Join Result:")
print(inner_join)

# Left Join: Returns all rows from the left DataFrame and matched rows from the right DataFrame
left_join = pd.merge(df1, df2, on='ID', how='left')
# Rows in df1 without a match in df2 will have NaN for columns from df2
print("\nLeft Join Result:")
print(left_join)

# Right Join: Returns all rows from the right DataFrame and matched rows from the left DataFrame
right_join = pd.merge(df1, df2, on='ID', how='right')
# Rows in df2 without a match in df1 will have NaN for columns from df1
print("\nRight Join Result:")
print(right_join)

# Outer Join: Returns all rows from both DataFrames, with NaN where there is no match
outer_join = pd.merge(df1, df2, on='ID', how='outer')
# This will include all rows from both DataFrames
print("\nOuter Join Result:")
print(outer_join)
```

```python
# Importing pandas library
import pandas as pd

# Loading the Churn_Modelling dataset
data = pd.read_csv('Churn_Modelling.csv')

# Creating two subsets of the dataset for demonstration purposes
df1 = data[['RowNumber', 'CustomerId', 'Surname', 'CreditScore']]
df2 = data[['CustomerId', 'Geography', 'Gender', 'Age']]

# Inner Join: Returns rows where there is a match in both DataFrames
inner_join = pd.merge(df1, df2, on='CustomerId', how='inner')
# This will merge rows with matching 'CustomerId' values
print("Inner Join Result:")
print(inner_join.head())

# Left Join: Returns all rows from the left DataFrame and matched rows from the right DataFrame
left_join = pd.merge(df1, df2, on='CustomerId', how='left')
# Rows in df1 without a match in df2 will have NaN for columns from df2
print("\nLeft Join Result:")
print(left_join.head())

# Right Join: Returns all rows from the right DataFrame and matched rows from the left DataFrame
right_join = pd.merge(df1, df2, on='CustomerId', how='right')
# Rows in df2 without a match in df1 will have NaN for columns from df1
print("\nRight Join Result:")
print(right_join.head())

# Outer Join: Returns all rows from both DataFrames, with NaN where there is no match
outer_join = pd.merge(df1, df2, on='CustomerId', how='outer')
# This will include all rows from both DataFrames
print("\nOuter Join Result:")
print(outer_join.head())
```
