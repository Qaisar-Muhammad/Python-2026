```python
# FUNCTION DEFINITION AND TYPES

# ============================================
# WHAT IS A FUNCTION?
# ============================================
# • A function is a reusable block of code that performs a specific task
# • Functions help organize code, reduce repetition, and enhance readability
# • Functions can take inputs (parameters) and return outputs (return values)
# • Functions are defined using the 'def' keyword in Python

# ============================================
# TYPES OF FUNCTIONS IN PYTHON
# ============================================

# 1. BUILT-IN FUNCTIONS
#    • Pre-defined functions in Python (print(), len(), max(), min(), type(), etc.)
#    • Readily available without need to define them
#    Example: print("Hello"), len([1, 2, 3])

# 2. USER-DEFINED FUNCTIONS
#    • Functions created by the programmer using the 'def' keyword
#    • Customized to perform specific tasks according to user requirements
#    Example: def greet(name): print(f"Hello {name}")

# 3. LAMBDA FUNCTIONS (Anonymous Functions)
#    • Small, anonymous functions defined using the 'lambda' keyword
#    • Usually used with map(), filter(), or sorted()
#    • Can have only one expression
#    Example: lambda x: x * 2

# 4. RECURSIVE FUNCTIONS
#    • Functions that call themselves to solve a problem
#    • Must have a base case to prevent infinite recursion
#    Example: factorial(n) calls factorial(n-1)

# ============================================
# KEY CHARACTERISTICS OF FUNCTIONS
# ============================================
# • PARAMETERS: Variables that function receives as input
# • ARGUMENTS: Actual values passed to the function
# • RETURN VALUE: Output that function sends back to caller
# • SCOPE: Region where variable is accessible (local or global)
# • DEFAULT PARAMETERS: Parameters with default values
# • *ARGS: Accept variable number of positional arguments
# • **KWARGS: Accept variable number of keyword arguments
```

```python
def sum_nat(n):
    num=0
    for i in range (1,n+1):
        num=num+i
    return  num 

sum_nat(5)
```

```python
def is_palindrome(text):
    cleaned = text.replace(" ", "").lower()
    left, right = 0, len(cleaned) - 1
    
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    return True
```
