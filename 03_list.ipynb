```python
# list
a=['ali', 'ahmed', 'hassan', 'ali' ]
print(type(a))
print(a)
```

```python
student_info=['ali', 25, 'male', 'computer science']
print(type(student_info))
print(student_info)
```

```python
print(len(a)) # 4
print(len(student_info)) # 4
```

```python
print(student_info[0]) # ali
print(student_info[1]) # 25 
print(student_info[2]) #
print(student_info[3]) # computer science
# print(student_info[4]) # IndexError: list index out of range
print(student_info[-1]) # computer science
print(student_info[-2])
print(student_info[-3])
print(student_info[-4]) 
```

```python
print (student_info[0:2]) # ['ali', 25]
print (student_info[1:3]) # [25, 'male']
print (student_info[2:4]) # ['male', 'computer science']
print (student_info[:2]) # ['ali', 25]      
print (student_info[2:]) # ['male', '   computer science']      
```

```python
#[[]]
print(student_info[0][0]) # a
print(student_info[0][1]) # l
print(student_info[0][2]) # i
```

```python
#[[]] different 
print(student_info[0][0]) # a
print(student_info[0][1]) # l
print(student_info[0][2]) # i

```

```python
# indexing and slicing in list
print(student_info[0]) # ali
print(student_info[0][0]) # a   
print(student_info[0][1]) # l
print(student_info[0][2]) # i 
# print(student_info[0][3]) # IndexError: string index out of range
print(student_info[0][0:2]) # ali
print(student_info[0][1:3]) # li    
print(student_info[0][2:4]) # i
print(student_info[0][:2]) # ali    
print(student_info[0][1:]) # li 
print(student_info[0][0:]) # ali
print(student_info[0][0:3]) # ali
```

```python
# list concatenation
list1 = [1, 2, 3]  
list2 = [4, 5, 6]
concatenated_list = list1 + list2
print(concatenated_list)  # Output: [1, 2, 3
```

```python
# string concatenation
str1 = "Hello, "
str2 = "world!" 
concatenated_str = str1 + str2
print(concatenated_str)  # Output: "Hello, world!"
```

```python
#membership operator in list
print('ali' in student_info) # True
print('ahmed' in student_info) # False
print(25 in student_info) # True
print('computer science' in student_info) # True
print('math' in student_info) # False   
```

```python
# mutability of list
# list is mutable in string
lists_subjects = ['math', 'physics', 'chemistry']
print(lists_subjects) # ['math', 'physics', 'chemistry']
lists_subjects[0] = 'biology'
print(lists_subjects) # ['biology', 'physics', 'chemistry']
#update
lists_subjects[1] = 'english'
print(lists_subjects) # ['biology', 'english', 'chemistry']
```

```python
#extend function in list
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)
print(list1) # [1, 2, 3, 4, 5, 6]
```

```python
#append function in list
list1 = [1, 2, 3]
list1.append(4)
print(list1) # [1, 2, 3, 4]
```

```python
#delete function in list
list1 = [1, 2, 3, 4]
list1.remove(2)
print(list1) # [1, 3, 4]
```

```python
#del with index
list1 = [1, 2, 3, 4]
del list1[1]
print(list1) # [1, 3, 4]
```

```python
#pop function in list   
list1 = [1, 2, 3, 4]
popped_element = list1.pop()
print(popped_element) # 4
```

```python
#sort function in list
list1 = [3, 1, 4, 2]
list1.sort()
print(list1) # [1, 2, 3, 4]
```

```python
#diff between sort and sorted function in list
list1 = [3, 1, 4, 2]
sorted_list = sorted(list1)
print(sorted_list) # [1, 2, 3, 4]
print(list1) # [3, 1, 4, 2]
#exlaination: sort function modifies the original list and returns None, while sorted function returns a new sorted list and does not modify the original list. 
```

```python
# shallow copy of list
list1 = [1, 2, 3]   
list2 = list1.copy()
print(list2) # [1, 2, 3]
```

```python
#deep copy of list
#all related to copy module
import copy
list1 = [1, 2, 3]
list2 = copy.deepcopy(list1)
print(list2) # [1, 2, 3]
print(list1) # [1, 2, 3]

```

```python
#split function in list
string = "Hello, world!"
list_of_words = string.split()
print(list_of_words) # ['Hello,', 'world!'] 
```

```python
#split method in string
string = "Hello, world!"    
print(string.split()) # ['Hello,', 'world!']
```

```python
#explain deep copy and shallow copy in list
#shallow copy creates a new list but does not create copies of the objects inside the list
#deep copy creates a new list and also creates copies of the objects inside the list
```
